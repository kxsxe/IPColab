```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/rice.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gImg, cmap ='gray')
cv2_imshow(gImg)
```
